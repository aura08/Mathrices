```python
# put them all together:
def inMatriceImage(img):
    luminanced=img[:,:,0]
    blur = cv2.GaussianBlur(luminanced,(5,5),0)
    ret3,th3 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    thresh=np.copy(th3)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    contoured=cv2.drawContours(th3, contours, -1, (0,255,0), 3)
    innerCnt=[]
    for cnt,hie in zip(contours,hierarchy[0]):
        if(hie[3]==0 and cv2.contourArea(cnt)>250):
            x,y,w,h=cv2.boundingRect(cnt)
            innerCnt.append(cnt)
            cv2.rectangle(contoured,(x,y),(x+w,y+h),(100,200,100),1)
    innerCnt=sorted(innerCnt,key=lambda cnt:cv2.boundingRect(cnt)[3],reverse=True)[:2]#sort by heigt
    Brackets=innerCnt[:2]
    numbers=innerCnt[2:]
    for Bracket in Brackets:
        x,y,w,h=cv2.boundingRect(Bracket)
        cv2.rectangle(th3,(x,y),(x+w,y+h),(200,200,200),10)
    left,right=[cv2.boundingRect(bracket) for bracket in Brackets]
    witdh=right[0]+right[2]
    height=left[3] if left[3]>right[3] else right[3]
    cv2.rectangle(th3,(left[0],left[1]),(witdh,left[1]+height),(200,200,200),10)
    cv2.drawContours(thresh,Brackets, -1, (255,255,255), 50)
    crop_img = thresh[left[1]:left[1]+height, left[0]:left[0]+witdh]
    return cv2.resize(crop_img,(720,720))

def parser(img):
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    numbers=[]
    centers=[]
    for cnt,hie in zip(contours,hierarchy[0]):
        if(hie[3]==0 and cv2.contourArea(cnt)>250):
            numbers.append(cv2.boundingRect(cnt))
    return sorted(numbers,key=lambda x:np.linalg.norm(np.array([x[0],x[1]])-np.array([0,0])))
    # sort for distance to orjin


    

    
```
